#Operator Aritmatika

```python
buku_tulis_himam = 55
buku_tulis_vico = 15
```

```python
#Penjumlahan
total_buku_tulis = buku_tulis_himam + buku_tulis_vico
print(total_buku_tulis)
```

```python
#Pengurangan

selisi_buku_tulis = buku_tulis_himam - buku_tulis_vico
print(selisi_buku_tulis)
```

```python
#Perkalian
jumlah_kardus_buku_tulis = 3
isi_buku_tulis_perkardus = 10

total_buku_tulis = jumlah_kardus_buku_tulis * isi_buku_tulis_perkardus

print(total_buku_tulis)
```

```python
#Pembagian
jumlah_anak_yatim = 9

jumlah_buku_tulis_tiap_anak = total_buku_tulis / jumlah_anak_yatim
print(jumlah_buku_tulis_tiap_anak)
```

```python
#Pembagian bulat
jumlah_anak_yatim = 9

jumlah_buku_tulis_tiap_anak = total_buku_tulis // jumlah_anak_yatim
print(jumlah_buku_tulis_tiap_anak)
```

```python
#Sisa bagi

sisa_buku_tulis = total_buku_tulis % jumlah_anak_yatim
print(sisa_buku_tulis)
```

```python
#Perpangkatan

sisi = 5
luas_persegi = sisi ** 2

print(luas_persegi)
```

```python
luas_persegi = 36

sisi = luas_persegi ** (1/2)
print(sisi)
```
